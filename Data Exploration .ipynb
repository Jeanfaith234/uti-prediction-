```python

#we see that the values are in strings so let us convert them to appropriate types
# we first convert the data to string then look for , character and remove it
df =df.astype('str')
df =df.apply(lambda x: x.str.upper())
```

## Encoding Categorical Data

```python
labels = df['HIV status\n(Positive/Negative/Unknown)'].astype('category').cat.categories.tolist()
replace_map_comp = {'HIV status\n(Positive/Negative/Unknown)' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}

print(replace_map_comp)
```

```python
df.replace(replace_map_comp, inplace=True)

print(df.head())
```
